```python
url = 'https://www.worldometers.info/coronavirus/'
```

```python
import requests
```

```python
html = requests.get(url).text
```

```python
type(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tbody = soup.find('tbody')
```

```python
tr = tbody.find_all('tr')
```

```python
tr[0]
```

```python
tr[1]
```

```python
tr[8]
```

```python
for item in range(8, len(tr)):
    print(tr[item])
```

```python
for item in range(8, len(tr)):
    td=(tr[item]).find_all('td')
    print(td)
```

```python
# [[1, 'usa', 55655, 5587], [2, 'india', 554, 888]]
```

```python
complete_data = []
for item in range(8, len(tr)):
    country_data = []
    table_row = tr[item]
    td = table_row.find_all('td') # one table row has multiple table data
    for item in td:
        country_data.append(item.text) # extract data of each country
    complete_data.append(country_data) # append all data of each country on a list
        
```

```python
clean_data = list(map(lambda x: x[1:10], complete_data))
```

```python
clean_data
```
