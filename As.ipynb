* BMI Calculator

````python
# without logical operator

user_input1 = float(input("Enter your weight (in kg): "))
user_input2 = float(input("Enter your height (in cm): "))

height_in_m = user_input2 / 100

BMI = user_input1 / height_in_m ** 2

if BMI < 18.5:
    print("You are underweight.")

elif BMI >= 18.5:
    print("You have normal weight.")

elif BMI <= 24.9:
    print("You have normal weight.")
    
elif BMI >= 25:
    print("You are overweight.")

elif BMI <= 29.9:
    print("You are overweight.")

else:
    print("Obesity.")
````

* WAP to print vowels and consonants in given word
```python
s = "aeyi"
```

````python
s = 'aeyi'

start = 0
end = len(s)
step = 1

for index in range(start, end, step):
    letter = s[index]
    
    if letter == 'a' or  letter == 'e' or letter == 'i'  or letter == 'o' or letter == 'u':
        print(f'Vowel = {letter}')
    
    else:
        print(f'Consonant = {letter}')
````

* WAP to count vowel and consonants in given string
```Python
s = 'HELLO'
```

````python
s = 'HELLO'

start = 0
end = len(s)
step = 1

vowel = 0
consonant = 0

for index in range(start, end, step):
    letter = s[index]
    
    if letter == 'A' or  letter == 'E' or letter == 'I'  or letter == 'O' or letter == 'U':
        print(f'{letter} is a vowel')
        vowel = vowel + 1
    
    else:
        print(f'{letter} is a consonant')
        consonant = consonant + 1

print(f'Total no. of vowels = {vowel}')
print(f'Total no. of consonants = {consonant}')
````

* WAP to count vowel and consonants in given string
