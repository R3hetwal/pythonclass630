```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
span = soup.find('h5')
```

```python
span
```

```python
import re

regex = re.compile(r'<[^>]+>')

def remove_html(string):
    return regex.sub('', string)

text = '<h5>As of : <span class="text-org" id="dDate">2022-12-07 15:00:00</span></h5>'
span = remove_html(text)
span = span.replace('As of : ', '')
print(f"{span}")
```

```python
span 
```

```python
tbody = soup.find('tbody')
```

```python
table = tbody.find_all('tr')
```

```python
tds = table[0].find_all('td')
```

```python
tds[3].text.lstrip()
```

```python
for item in range(1, len(table)):
    print(table[item])
```

```python
for item in range(1, len(table)):
    tds=(table[item]).find_all('td')
    print(tds)
```

```python
complete_data = []
for item in range(1, len(table)):
    company_data = []
    table_row = table[item]
    tds = table_row.find_all('td') # one table row has multiple table data
    for item in tds:
        company_data.append(item.text) # extract data of each company
   
    complete_data.append(company_data) # append all data of each company on a list
```

```python
clean_data = list(map(lambda x: x[1:10], complete_data))
```

```python
clean_data
```
