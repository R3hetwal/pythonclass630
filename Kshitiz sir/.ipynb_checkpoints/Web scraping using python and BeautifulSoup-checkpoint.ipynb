```python
url = 'https://www.worldometers.info/coronavirus/'
```

```python
import requests
```

```python
html = requests.get(url).text
```

```python
type(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tbody = soup.find('tbody')
```

```python
tr = tbody.find_all('tr')
```

```python
tr[8]
```

```python
for item in range(8, len(tr)):
    print(tr[item])
    
```

```python
# [[1,'usa', 55655, 5587], [2, 'inida', 554, 888]]
```

```python
complete_data = []
for item in range(8, len(tr)):
    country_data = []
    table_row = tr[item]
    td = table_row.find_all('td') # one table row has multiple table data
    for item in td:
        country_data.append(item.text) # extract data of each country
        
    complete_data.append(country_data) # append all data of each country on a list
    
```

```python
complete_data[0]
```

```python
clean_data = list(map(lambda x: x[1:10], complete_data ))
```

```python
clean_data
```

```python
# ! pip install pandas
# ! pip install matplotlib
# ! pip install seaborn 
```

```python
import pandas as pd # as pd is alias name convention
```

```python
df = pd.DataFrame(clean_data)
```

```python
df
```

```python
col_names = ['country',
             'total cases',
             'new cases', 
             'total death',
             'new death', 
             'total recovered',
            'new recovered', 
            'active cases', 
            'critital']
```

```python
# updating column name on pandas dataframe
```

```python
df.columns = col_names # update column name of a dataframe 
```

```python
df.head() # print 5 data rows from top
```

```python
df.tail() # last 5 data 
```

# convert dataframe to csv

```python
df.to_csv('covid-live_dec7.csv', index = False) # index = False -- dont save index
```
